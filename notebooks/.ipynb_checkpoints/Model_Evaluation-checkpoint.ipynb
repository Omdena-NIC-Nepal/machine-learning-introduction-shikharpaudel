```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# machine learning env
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import joblib
```

```python
# load preprocessed data
X_train = pd.read_pickle("X_train.pkl")
X_test = pd.read_pickle("X_test.pkl")
y_train = pd.read_pickle("y_train.pkl")
y_test = pd.read_pickle("y_test.pkl")
```

```python
# load the trained ridge model
best_ridge = joblib.load("ridge_model.pkl")
```

```python
# evaluate the model
# train a new regression model
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
```

```python
# predictions
y_pred_lr = lr_model.predict(X_test)
y_pred_ridg = best_ridge.predict(X_test)
```

```python
# metrics
mse_lr = mean_squared_error(y_test,y_pred_lr)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test,y_pred_lr)
mse_ridge = mean_squared_error(y_test,y_pred_ridg)
mae_ridge = mean_absolute_error(y_test,y_pred_ridg)
r2_ridge = r2_score(y_test,y_pred_ridg)
```

```python
print("Linear Regression Evaluation:")
print(f"Mean Squared Error: {mse_lr:.2f}")
print(f"Mean Absolute Error: {mae_lr:.2f}")
print(f"r2 Score : {r2_lr:2f}")
print("\nRidged Regression Evaluation")
print(f"Mean Squared Ridge: {mse_ridge:.2f}")
print(f"Mean Absolute Error Ridge: {mae_ridge:.2f}")
print(f"r2 Score Ridge: {r2_ridge:.2f}")
```

```python
# plot residuels
residuels_lr = y_test - y_pred_lr
residuels_ridge = y_test - y_pred_ridg
```

```python
plt.figure(figsize = (12,5))
# linear regression resedual
plt.subplot(1,2,1)
plt.scatter(y_pred_lr, residuels_lr,color ='blue', alpha = 0.5)
plt.axhline(y = 0, color = 'red',linestyle = '--')
plt.title("Linear Regression Residuels", pad = 20)
plt.xlabel("Predicted Values")
plt.ylabel("Residuels")
# risge regression residuels
plt.subplot(1,2,2)
plt.scatter(y_pred_ridg, residuels_ridge, color = 'blue', alpha = 0.5)
plt.axhline(y = 0, color = 'red', linestyle = '--')
plt.title("Ridged Regression Residuels", pad = 20)
plt.xlabel("Predicted Values")
plt.ylabel("Residuels")
plt.tight_layout()
plt.show()
```

```python
# distribution of residuals
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(residuels_lr, kde=True, color='blue')
plt.title("Linear Regression Residuals Distribution")

plt.subplot(1, 2, 2)
sns.histplot(residuels_ridge, kde=True, color='green')
plt.title("Ridge Regression Residuals Distribution")
plt.tight_layout()
plt.show()
```

```python
# compare with a reduced feature 
reduced_features = ['rm', 'lstat']
X_train_reduced = X_train[reduced_features]
X_test_reduced = X_test[reduced_features]
```

```python
# Train and evaluate Linear Regression with reduced features
lr_reduced = LinearRegression()
lr_reduced.fit(X_train_reduced, y_train)
y_pred_reduced = lr_reduced.predict(X_test_reduced)

mse_reduced = mean_squared_error(y_test, y_pred_reduced)
r2_reduced = r2_score(y_test, y_pred_reduced)
```

```python
print("\nLinear Regression with Reduced Features (rm, lstat):")
print(f"Mean Squared Error: {mse_reduced:.2f}")
print(f"R-squared: {r2_reduced:.2f}")
```
