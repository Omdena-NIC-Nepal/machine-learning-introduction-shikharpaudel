```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("BostonHousing.csv")
data.head()
```

```python
# correlation heatmap
plt.figure(figsize = (10,8))
sns.heatmap(data.corr(), annot = True, cmap = 'coolwarm',vmin = -1, vmax = 1,center = 0, square = True, linewidths = 0.5)
plt.title("Correlation Heatmap", pad = 20)
plt.show()
```

```python
# Scatter plots feature Vs target
features = ['rm', 'lstat', 'ptratio', 'tax', 'indus']
fig, axes = plt.subplots(2,3, figsize = (18,12))
for i , feature in enumerate(features):
    ax = axes[i//3, i%3]
    sns.scatterplot(data = data, ax = ax, x = feature, y = 'medv')
    ax.set_title(f"{feature} vs MEDV")
fig.delaxes(axes[1,2])
plt.show()
```

```python
# box plot to detect outliers
plt.figure(figsize = (12,10))
sns.boxplot(data = data)
plt.title("Boxplot of Features to Identify Outliers", pad = 20)
plt.show()
```
